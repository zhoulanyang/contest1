```python
#coding:utf-8
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, skew
from sklearn.preprocessing import StandardScaler
plt.rcParams['font.sans-serif'] = ['SimHei']
sns.set_style('whitegrid',{'font.sans-serif':['simhei','Arial']})
train_file = '../train_dataset/train_dataset.csv'
test_file = '../test_dataset/test_dataset.csv'

train = pd.read_csv(train_file, encoding='utf-8')

test = pd.read_csv(test_file, encoding='utf-8')
print('训练集的size为:{}'.format(train.shape))
print('测试集的size为:{}'.format(test.shape))
y = train['信用分']

from scipy.special import boxcox

lam = 6
train[u'信用分'] = boxcox(train['信用分'], lam)
# train.drop(u'信用分', axis=1, inplace=True)
# print(train.shape)
print(skew(train['信用分']))
sns.distplot(train['信用分'], fit=norm)
plt.show()
```

```python
# sns.distplot(np.sqrt(train[u'用户账单当月总费用（元）']), fit=norm)
# 用户账单当月总费用（元）做特征处理
lam = 0.5 
train[u'用户账单当月总费用（元）'] = boxcox(train[u'用户账单当月总费用（元）'], lam)
sns.distplot(train[u'用户账单当月总费用（元）'], fit=norm)
plt.show()
print(skew(train[u'用户账单当月总费用（元）']))
```

```python
# sns.distplot(np.sqrt(train['用户当月账户余额（元）']), fit=norm)

lam = -0.01
train['用户当月账户余额（元）'] = boxcox(train['用户当月账户余额（元）'], lam)
sns.distplot(train['用户当月账户余额（元）'], fit=norm)
plt.show()
print(skew(train['用户当月账户余额（元）']))
```

```python
lam = 2.5
train[u'用户近6个月平均消费值（元）'] = np.log1p(train[u'用户近6个月平均消费值（元）'])
train[u'用户近6个月平均消费值（元）'] = boxcox(train[u'用户近6个月平均消费值（元）'], lam)
sns.distplot(train[u'用户近6个月平均消费值（元）'], fit=norm)
print(skew(train[u'用户近6个月平均消费值（元）']))
plt.show()
```

```python
var = '用户网龄（月）'
# sns.distplot(train[var], fit=norm)
lam = 0.8

train[var] = boxcox(train[var], lam)
print("skewness of 用户网龄 is :{}".format(skew(train[var])))

sns.distplot(train[var], fit=norm)
plt.show()
```

```python
train_Id = train[u'用户编码']
test_Id = test[u'用户编码']

all_data = pd.concat((train, test)).reset_index(drop=True)
print('The Initial size of all_data is :{}'.format(all_data.shape))
all_data.drop(u'用户编码', axis=1, inplace=True)
all_data.drop(u'信用分', axis=1, inplace=True)
print('All data size is :{}'.format(all_data.shape))
all_data.head()
```

```python
numeric_feats = all_data.dtypes[all_data.dtypes != "object"].index

# Check the skew of all numerical features
skewed_feats = all_data[numeric_feats].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
print("\nSkew in numerical features: \n")
skewness = pd.DataFrame({'Skew' :skewed_feats})
skewness.head(10)
```

```python
skewness = skewness[abs(skewness) > 0.75]
print("There are {} skewed numerical features to Box Cox transform".format(skewness.shape[0]))

from scipy.special import boxcox1p
skewed_features = skewness.index
lam = 0.6
for feat in skewed_features:
    #all_data[feat] += 1
    all_data[feat] = boxcox1p(all_data[feat], lam)
```

```python
all_data = pd.get_dummies(all_data)
print(all_data.shape)
```
